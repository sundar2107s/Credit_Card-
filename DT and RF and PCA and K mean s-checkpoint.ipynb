```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
data=pd.read_excel("Cleaned_Credit.xlsx")
df=pd.DataFrame(data)
df.columns

```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
X=df.drop('Df',axis=1).drop('x12',axis=1)
y=df['Df']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
predictions = dtree.predict(X_test)

from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(y_test,predictions)
plt.plot(fpr, tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC curve for Default classifier')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.grid(True)
print(metrics.roc_auc_score(y_test, predictions))
```

```python
#PCA(Princeple component Analysis)
#First we need to standardize the data to perform a PCA
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
attributes=df[['x1','x2','x3','x4','x5','x6','x7','x8','x9','x10','x11']]
scaler=StandardScaler()
scaler.fit(attributes)
scaled_data=scaler.transform(attributes)
scaled_data1=pd.DataFrame(scaled_data)
scaled_data1.info()

```

```python
scaled_data1.head()
```

```python
pca=PCA(n_components=5)
pca.fit(scaled_data1)
x_pca=pca.transform(scaled_data1)
plt.plot(pca.explained_variance_ratio_)
plt.xlabel("Number of Components")
plt.ylabel("Variance_Ratio")
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x=x_pca[:,0],y=x_pca[:,1],data=scaled_data1,c=df['Df'])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
#sns.lmplot(x=scaled_data1[0],y=scaled_data1[1],data=scaled_data1,hue=df['Df'],fit_reg=False)
```

```python
pca.components_
len(pca.components_)
```

```python
df_comp=pd.DataFrame(pca.components_,columns=['x1','x2','x3','x4','x5','x6','x7','x8','x9','x10','x11'])
df_comp
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap="plasma")
```
